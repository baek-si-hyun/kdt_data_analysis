```python
import pandas as pd
# 데이터 확인

happiness_df = pd.read_csv('./datasets/happiness_report_2023.csv')
display(happiness_df)
```

```python
# 칼럼 변경
happiness_df.columns = ['country','region', 'happiness', 'gdp', 'support', 'health', 'freedom', 'generosity', 'corruption']
happiness_df
```

```python
happiness_df.index += 1
happiness_df
```

```python
happiness_df['region'].value_counts()
```

```python
# 결측치 검사
happiness_df.isna().sum()
```

```python
# 결측치 확인
happiness_df[happiness_df['health'].isna()]
```

```python
display(happiness_df.describe().T)
```

```python
# 결측치 평균값으로 채우기
happiness_df['health'] = happiness_df['health'].fillna(happiness_df['health'].mean())
```

```python
# 결측치 재확인
happiness_df.isna().sum()
```

```python
# 데이터 요약 정보 확인
happiness_df.info()
happiness_df.dtypes
```

```python
# 정상치 범위 확인
import numpy as np
from scipy.stats import iqr

happiness_Q1 = happiness_df.describe().T.loc['gdp', '25%']
happiness_Q3 = happiness_df.describe().T.loc['gdp', '75%']
print(happiness_Q1, happiness_Q3)

iqr_value = iqr(happiness_df.gdp)
print(iqr_value)

lower_bound = (happiness_Q1 - 1.5 * iqr_value)
upper_bound = (happiness_Q3 + 1.5 * iqr_value)
print(f'정상치 범위: {lower_bound} ~ {upper_bound}')
```

```python
# 이상치 데이터 확인
happiness_df[(happiness_df['gdp'] < lower_bound) | (happiness_df['gdp'] > upper_bound)]
```

```python
# 이상치 데이터 제거
happiness_df = happiness_df[~(happiness_df['gdp'] < lower_bound) | (happiness_df['gdp'] > upper_bound)]
happiness_df
```

```python
gdp_mean_df = happiness_df.groupby('region').mean('gdp').reset_index()
display(gdp_mean_df)

gdp_min_df = happiness_df.groupby('region').min('gdp').reset_index()
display(gdp_min_df)

gdp_max_df = happiness_df.groupby('region').max('gdp').reset_index()
display(gdp_max_df)

gdp_std_df = happiness_df.groupby('region')['gdp'].std().reset_index()
display(gdp_std_df)

hp_mean_df = happiness_df.groupby('region').mean('happiness').reset_index()
display(gdp_mean_df)

hp_min_df = happiness_df.groupby('region').min('happiness').reset_index()
display(gdp_min_df)

hp_max_df = happiness_df.groupby('region').max('happiness').reset_index()
display(gdp_max_df)

hp_std_df = happiness_df.groupby('region')['happiness'].std().reset_index()
display(hp_std_df)
```

### 2023 지역별 GDP에 따른 행복지수 분석 보고서
#### 서구 유럽 (Western Euripe)

    서구 유럽 지역은 경제적으로 안정된 국가들로 구성되어 있다. 
    GDP의 평균은 약 1.911이며, 표준편차는 0.188로 매우 낮다. 
    이는 서구 유럽 국가들의 경제력이 평균적으로 안정되어 있고, GDP 수준이 비교적 일관되게 분포되어 있음을 말한다. 
    또한 행복지수의 평균은 6.894로 안정적이며, 표준편차는 0.558로 GDP와는 별개로 일부 변동성을 보인다. 
    
#### 중동 및 북아프리카 (Middle East and North Africa)

    중동 및 북아프리카 지역은 경제적으로 다양한 국가들로 구성되어 있다. 
    GDP의 평군은 1.151이며, 표준편차는 0.285로 서구 유럽에 비해 높은 편이다. 
    이는 중동 및 북아프리카 국가들 간의 GDP 수준이 상당히 다양하게 분포되어 있음을 말한다. 
    또한, 행복지수의 평균은 5.102로 낮으며, 표준편5차는 1.264로 매우 높은 편이다. 
    이는 GDP 수준의 양극화가 행복지수에도 영향을 미치고 있음을 나타낸다.
    
#### 남아시아 (South Asia)

    남아시아 지역은 경제적으로 다양한 국가들로 구성되어 있다. 
    GDP의 평균은 약 1.070이며, 표준편차는 0.255로 중동 및 북아프리카에 비해 낮은 편이다. 
    하지만 행복지수의 편균은 4.089로 낮으며, 표준편차는 1.181로 매우 높은 편이다. 
    이는 남아시아 국가들 사이에 행복지수의 큰 차이가 존재하고 있음을 시사한다.
    
#### 북미와 호주, 뉴질랜드 (North Ameri)

    북미와 호주, 뉴질랜드 지역은 경제적으로 안정된 국가들로 구성되어 있다. 
    GDP의 평균은 약 1.900이며, 표준편자는 0.058로 매우 낮다. 
    이는 이 지역의 국가들이 경제적으로 평균적으로 안정되어 있으며, GDP 수준이 비교적 고르게 분포되어 있음을 말한다. 
    또한, 행복지수의 평균은 7.018로 높으며, 표준편차는 0.019로 상대적으로 낮은 편이다. 
    이는 북미와 호주, 뉴질랜드 국가들이 경제적 안정과 행복지수의 높은 수준을 함께 유지하고 있다는 것을 나타낸다.
    
#### 결론

    지역별 GDP 수준에 따른 행복지수의 변동을 분석한 결과, 경제적으로 안정된 지역들이 일반적으로 행복지수의 변동이 적은 반면, 
    경제적으로 어려움을 겪고 있는 지역들은 행복지수의 변동성이 더 큰 경향을 보인다.
    는 것을 나타냅니다.역들은 행복지수인다.
