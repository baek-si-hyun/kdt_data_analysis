### Numpy Task

```python
import numpy as np
# 1~10까지의 요소를 ndarray에 담는다.
ndarray1 = np.arange(1, 11)
print(ndarray1)

# 3행2열 list를 ndarray에 담고 각 요소는 순서대로 6, 5, 4, 3, 2, 1이다
ndarray1 = np.arange(1, 7)
ndarray2 = ndarray1.reshape((3,2))
print(ndarray2[::-1,::-1])
#------------------------------------------------
ndarray1 = np.array([[6, 5], [4, 3], [2, 1]])
print(ndarray1)

# 4행 1열 ndarray, 순서대로 10, 8, 6, 4, 를 담는다
ndarray1 = np.arange(start=4, stop=11, step=2)[::-1]
print(ndarray1)
#------------------------------------------------
ndarray1 = np.array([10, 8, 6, 4])
print(ndarray1)

# 1 ~ 10까지 요소를 ndarray에 담고 각 요소에 5식 더해준다.
ndarray1 = np.arange(1, 11)
ndarray1 = np.array([i + 5 for i in ndarray1])
print(ndarray1)
#------------------------------------------------
ndarray1 = np.arange(1, 11) + 5
print(ndarray1)

# axis 0이 10인 shape이면서 모든 원소가 0, dtype은 int32인 ndarray만들기
ndarray1 = np.zeros((10,),dtype=np.int32)
print(ndarray1)
#------------------------------------------------
ndarray1 = np.zeros((10,), dtype=np.int32)
print(ndarray1)

# axis 0이 3, axis1이 4인 shape이면서 모든 원소가 1인 ndarray 만들기
ndarray2 = np.ones((3, 4),dtype=np.int32)
print(ndarray2)
#------------------------------------------------
ndarray2 = np.ones((3, 4), dtype=np.int32)
print(ndarray2)

# axis 0이 5, 각 요소가 0~4인 ndarray 만들기
ndarray1 = np.arange(5,) % 5
print(ndarray1)
#------------------------------------------------
ndarray1 = np.arange(5) % 5
print(ndarray1)

# 아래의 ndarray1이 주어졌을 때 순서대로 문제를 해결한다.
ndarray1 = np.arange(start=0, stop=16)
# 2 Dimension, axis 1은 2로 변경
# 2 Dimension, axis 0은 8로 변경
ndarray2 = ndarray1.reshape((8, 2))
print(ndarray2)
#------------------------------------------------
ndarray2 = ndarray1.reshape((8, 2))
print(ndarray2)

# 3 Dimension으로 변경
ndarray3 = ndarray1.reshape((8, 2, -1))
print(ndarray3)
#------------------------------------------------
ndarray3 = ndarray1.reshape((8, 2, -1))
print(ndarray3)

# ndarray3을 axis 1이 1인 2차원 ndarray로 변환
ndarray3 = ndarray1.reshape((-1, 1))
print(ndarray3)
#------------------------------------------------
ndarray3 = ndarray3.reshape((-1, 1))
print(ndarray3)

# ndarray3을 1 Dimension으로 변환
ndarray3 = ndarray1.reshape((-1,))
print(ndarray3)
#------------------------------------------------
ndarray3 = ndarray3.flatten()
print(ndarray3)

# 1차원 ndarray 1~100까지 값 중 97값 추출하기
ndarray1 = np.arange(1, 101)
print(ndarray1[-4])
#------------------------------------------------
ndarray1 = np.arange(1, 101)
print(ndarray1[96])

# 아래의 ndarray1이 주어졌을 때 순서대로 문제를 해결한다.
ndarray1 = np.arange(start=1, stop=21)
# 2차원 배열 axis 0은 4, axis 1은 5로 1~20 요소 값 중 15값 추출하기
ndarray2 = ndarray1.reshape((4, 5))
print(ndarray2[2, -1])
#------------------------------------------------
ndarray2 = ndarray1.reshape((4, 5))
print(ndarray2[2, 4])

# 2 Dimension, axis 0은 5로 변경
ndarray2 = ndarray2.reshape((5, -1))
print(ndarray2)
#------------------------------------------------
ndarray2 = ndarray2.reshape((5, -1))
print(ndarray2)

# 2행의 4번째 숫자 출력
print(ndarray2[1, 3])
#------------------------------------------------
print(ndarray2[1, 3])

# 1~100중 짝수만 출력
ndarray1 = np.arange(1, 101)
ndarray1 = ndarray1[ndarray1 % 2 == 0]
print(ndarray1)
#------------------------------------------------
ndarray1 = np.arange(2, 101, 2)
print(ndarray1)

# 위에서 구한 짝수들을 axis 0이 10인 2차원 배열로 변환 후 82~100까지 추출
ndarray2 = ndarray1.reshape((10, -1))
print(ndarray2)
ndarray2 = ndarray2[-2:]
print(ndarray2)
#------------------------------------------------
ndarray2 = ndarray1.reshape((10, -1))
print(ndarray2[-2:, 1:])

# 위에서 구한 요소 중 (86, 100) 좌표 구하기 -----
print(ndarray2[0][2])
print(ndarray2[1][-1])
#------------------------------------------------
print(ndarray2[-1, -1], ndarray2[-2, -3])

# 아래의 ndarray1이 주어졌을 때 순서대로 문제를 해결한다.
ndarray1 = np.arange(start=1, stop=101)
# 각 요소중 1의 자리수가 2인 수들만 추출하기
ndarray1 = ndarray1[ndarray1 % 10 == 2]
print(ndarray1)
#------------------------------------------------
ndarray1 = ndarray1[ndarray1 % 10 == 2]
print(ndarray1)

# 1~49요소 중 2와 5의 공배수 추출하기
ndarray1 = np.arange(1, 50)
ndarray1 = ndarray1[(ndarray1 % 2 == 0) & (ndarray1 % 5 == 0)]
print(ndarray1)
#------------------------------------------------
ndarray1 = np.arange(1, 50)
ndarray1 = ndarray1[(ndarray1 % 2 == 0) & (ndarray1 % 5 == 0)]
print(ndarray1)

# 위에서 추출한 공배수 중 20이상인 값만 추출하기
ndarray1 = ndarray1[ndarray1 >= 20]
print(ndarray1)
ndarray1 = [i for i in ndarray1 if i >= 20]
print(ndarray1)
#------------------------------------------------


# 데이터를 분석하여, 각 수치별 오름차순 및 내림차순 후 이름을 출력하세요.
# [[과일이름], [가격]], 가격별 내림차순
array1 = [['Mango', 'Apple', 'Pear', 'Pitch', 'Melon'], [1500, 1800, 2000, 2500, 8500]]
arr1 = np.array(array1[0])
arr2 = np.array(array1[1])
sorted_arr = np.argsort(arr2)
print(arr1[sorted_arr])
print(arr1[sorted_arr][::-1])
#------------------------------------------------
array1 = [['Mango', 'Apple', 'Pear', 'Pitch', 'Melon'], [1500, 1800, 2000, 2500, 8500]]
arr1 = np.array(array1[0])
arr2 = np.array(array1[1])
sorted_indices = np.argsort(arr2)[::-1] 
print(arr1[sorted_indices])

# [[이름], [나이], [학점]], 나이별 내림차순, 학점별 오름차순
array2 = [['둘리', '또치', '길동', '도너', '마이콜'], [7, 9, 40, 5, 20], ['C', 'A', 'B', 'F', 'D']]
arr1 = np.array(array2[0])
arr2 = np.array(array2[1])
arr3 = np.array(array2[2])
sorted = arr2.argsort()
print(arr1[sorted])
sorted = arr3.argsort()
print(arr1[sorted][::-1])
#------------------------------------------------
array2 = [['둘리', '또치', '길동', '도너', '마이콜'], [7, 9, 40, 5, 20], ['C', 'A', 'B', 'F', 'D']]
arr1 = np.array(array2[0])
arr2 = np.array(array2[1])
arr3 = np.array(array2[2])
age_sorted_indices = np.argsort(arr2)[::-1] 
grade_sorted_indices = np.argsort(arr3)  
print(arr1[age_sorted_indices])
print(arr1[grade_sorted_indices])

# [[국가이름], [GDP 성장률]], GDP 성장률별 오름차순
array3 = [['한국', '영국', '러시아', '남아프리카 공화국', '미국'], ['1.7%', '-0.6%', '0.3%', '1.2%', '1.4%']]
arr1 = np.array(array3[0])
arr2 = np.array(array3[1])
sorted = arr2.argsort()
print(arr1[sorted][::-1])
#------------------------------------------------
array3 = [['한국', '영국', '러시아', '남아프리카 공화국', '미국'], ['1.7%', '-0.6%', '0.3%', '1.2%', '1.4%']]
arr1 = np.array(array3[0])
arr2 = np.array(array3[1])
growth_rate_sorted_indices = np.argsort(arr2)  
print(arr1[growth_rate_sorted_indices])
```
