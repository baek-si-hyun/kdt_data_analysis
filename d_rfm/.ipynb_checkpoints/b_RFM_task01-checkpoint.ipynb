### RFM 분석 실습
- Recency: 얼마나 최근에 구매했는가
- Frequency: 얼마나 자주 구매했는가
- Monetary: 얼마나 많은 금액을 지출했는가

#### 특정 고객을 대상으로 회원 등급을 선정하고 높은 회원 등급을 가진 고객에게는 혜택을 제공하여 총성심있는 고객으로 유지 할 수 있는 전략을 세울수 있다.
#### RFM을 통해 사용자 특성 별로 다른 정책을 적용하고 서비스를 더 잘 사용하도록 유도하는 전략을 세울 수 있다.

#### 데이터 전처리

- PurchaseDate를 사용해 Recency로 변환
- CustomerID를 사용해 총 몇번 결제했는지를 카운트하여 Frequency로 변환- CustomerID 과 TransactionAmount를 이용하여지출 금액 관련 데이터를해하여 Monetary로 변환

```python
from datetime import datetime

# 데이터프레임의 PurchaseDate 열을 datetime 형식으로 변환
cp_df['PurchaseDate'] = pd.to_datetime(cp_df['PurchaseDate'])

# 현재 날짜와 시간 가져오기
current_datetime = pd.Timestamp(datetime.now())

# 각 행의 PurchaseDate와 현재 날짜 간의 차이를 계산하여 새로운 열에 추가
cp_df['days_pass_by'] = (current_datetime - cp_df['PurchaseDate']).dt.days
cp_df
```

```python
# cp_df = cp_df.groupby('CustomerID').reset_index()
# cp_df
```

```python
cp_dff = cp_df.groupby('CustomerID')['TransactionAmount'].sum().reset_index(name='Monetary')
display(cp_dff)
```

```python
cp_dff['Recency'] = cp_df.groupby('CustomerID')['days_pass_by'].min().reset_index()['days_pass_by']
display(cp_dff)
```

```python
cp_dff['Frequency'] = cp_df.groupby('CustomerID')['OrderID'].count().reset_index()['OrderID']
cp_rfm_df = cp_dff.drop(columns=['CustomerID'], axis=1)
display(cp_rfm_df)
```

```python
# 정규화
from sklearn.preprocessing import MinMaxScaler

normalization = MinMaxScaler()
rfm_normalization = normalization.fit_transform(cp_rfm_df)
```

```python
cp_rfm_df = pd.DataFrame(rfm_normalization, columns=cp_rfm_df.columns)
cp_rfm_df
```

```python
cp_dff['Location'] = cp_df.groupby('CustomerID')['Location'].apply(lambda x: ', '.join(x)).reset_index()['Location']

display(cp_dff)
```

```python
cp_dff['Product'] = cp_df.groupby('CustomerID')['ProductInformation'].apply(lambda x: ', '.join(x)).reset_index()['ProductInformation']

display(cp_dff)
```

```python
cp_dff['MeanPrice'] = cp_df.groupby('CustomerID')['TransactionAmount'].mean().reset_index()['TransactionAmount']
display(cp_dff)
```

```python
cp_dff['TotalScore'] = cp_dff['Recency'] + cp_dff['Frequency'] + cp_dff['Monetary']
cp_dff
```

```python
import numpy as np

l1, l2, l3, l4 = np.percentile(cp_dff['TotalScore'], [20, 40, 70, 90])
print(l1, l2, l3, l4)
```

```python
def get_level(x):
    if x <= l1:
        return 5
    if x > l1 and x <= l2:
        return 4
    if x > l2 and x <= l3:
        return 3
    if x > l3 and x <= l4:
        return 2
    return 1
```

```python
cp_dff['Level'] = cp_dff['TotalScore'].apply(get_level)
cp_dff['Level'] = cp_dff['Level'].replace([5, 4, 3, 2, 1], ['Iron', 'Bronze', 'Silver', 'Gold', 'Diamond'])
cp_dff
```

```python
print(cp_dff['Level'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

order_columns = ['Iron', 'Bronze', 'Silver', 'Gold', 'Diamond']

sns.countplot(x='Level', data=cp_dff, palette='muted', order=order_columns)
```

```python
fig, axes = plt.subplots(1, 1, figsize=(20, 8))

sns.boxplot(ax=axes, x='Level', y='MeanPrice', data=cp_dff, palette='muted', order=order_columns)
```

```python
cp_dff['Locations'] = cp_dff['Location'].str.split(', ')

cp_dff['Products'] = cp_dff['Product'].str.split(', ')

fig, axes = plt.subplots(1, 2, figsize=(20, 8))

sns.countplot(ax=axes[0], x='Products', hue='Level', data=cp_dff.explode('Products'), palette='muted', hue_order=order_columns)
sns.countplot(ax=axes[1], x='Locations', hue='Level', data=cp_dff.explode('Locations'), palette='muted', hue_order=order_columns)
```

### 마케팅 전략 제안
> 등급의 고객의 총 구매 금액이 낮은 'Paris'지역을 대상으로 필요한 상품들을 조사하여 구매욕을 높일 수 있습니다.
> 모든 지역에서 높은 구매 횟수를 보여주고 있는 'Silver' 등급의 고객들을 대상으로는 회원 등급 상승, 혜택 제공을 하는 것이 효과적일 수 있다.
> 모든 등급에서 재구매가 높은 Product C 상품과 관련있거나 유사한 상품을 더 늘림으로써 고객들의 관심을 유발할 수 있다.
> 회원수가 제일 많은 'Silver' 회원들을 타켓으로한 상품 출시를 기획하는 것이 효과적일수 있다.
> 총 결제 금액이 제일 높은 'Diamond' 회원들을 늘리기 위해 'Diamond'등급의 회원들만 즐길 수 있는 이벤트를 기획하여 'Diamond'의 등급의 차별화를 강조하는 것이 효과 적일 수 있다.

```python
import matplotlib.pyplot as plt
import seaborn as sns

cp_dff['Products'] = cp_dff['Product'].str.split(', ')

cp_dff = cp_dff.explode('Products')

plt.figure(figsize=(10, 6))
sns.boxplot(x='Products', y='TotalScore', data=cp_dff)
plt.title('Boxplot of TotalScore by Product')
plt.xlabel('Product')
plt.ylabel('TotalScore')
plt.show()
```
